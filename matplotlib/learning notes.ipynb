```python
from dateutil import parser as dateparser
import datetime
import pandas as pd
```

```python
dateparser.parse(u'2002, 04, 01')
```

```python
date1_s = '2016-05-01'
date2_s = '2016-08-01'
date1, date2 = [dateparser.parse(d) for d in [date1_s, date2_s]]
delta_date = datetime.timedelta(days=3)
date_range=mpl.dates.drange(date1, date2, delta_date)
y = np.random.randn(len(date_range))
fig_time = plt.figure()
ax = fig_time.add_subplot(111)
ax.plot_date(date_range, y, 'or:')
plt.show()
```

```python
def zh_dateformatter(d, pos=None):
    date = mpl.dates.num2date(d)
    timestamp = pd.Timestamp(date)
    y = timestamp.year
    m = timestamp.month
    d = timestamp.day
    return '{0}/{1}/{2}'.format(y%100, m, d)
```

```python
print zh_dateformatter(mpl.dates.date2num(dateparser.parse('2012-01-02')), None)
```

```python
ax.xaxis.set_major_formatter(mpl.ticker.FuncFormatter(zh_dateformatter))
ax.xaxis.set_major_locator(mpl.dates.DayLocator(interval=10))
ax.xaxis.set_minor_locator(mpl.dates.HourLocator(interval=48))
plt.show()
```

```python
fig_contour = plt.figure()
ax = fig_contour.add_subplot(211)
x = np.random.randn(50) + 2
y = np.random.randn(50) + 10
X, Y = np.meshgrid(x, y)
Z = X**2/4 + Y**2/16 - 5
contour_p = ax.contourf(Z)
plt.clabel(contour_p)
plt.colorbar(contour_p)
ax.set_aspect('equal')
plt.show()
```

```python
ax2 = fig_contour.add_subplot(212)
img = ax2.imshow(Z)
plt.colorbar(img)
plt.show()
```
