```python
from zipline.api import order, sid, symbol

class BuyGoog(zp.TradingAlgorithm):
    trace=False
    
    def __init__(self, trace=False):
        BuyGoog.trace=trace
        super(BuyGoog, self).__init__()
        
    def initialize(context):
        if BuyGoog.trace:
            print '---> initialize'
            print context
            print '<--- initialize'
    
    def handle_data(self, context):
        if BuyGoog.trace:
            print '---> handle_data'
            print context
        self.order(symbol('GOOG'), 1)
        if BuyGoog.trace:
            print '<--- handle_data'
```

```python
result = BuyGoog(trace=True).run(data['2013-01-01':'2016-01-01'])
```
