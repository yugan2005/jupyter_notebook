```python
import numpy as np
import matplotlib.pyplot as plt

# Try to reproduce the figure shown in images/exercise_1-1.png

# Our data...
x = np.linspace(0, 10, 100)
y1, y2, y3 = np.cos(x), np.cos(x + 1), np.cos(x + 2)
names = ['Signal 1', 'Signal 2', 'Signal 3']

fig, axes = plt.subplots(3)
for ax, y, name in zip(axes, [y1, y2, y3], names):
    ax.plot(x, y)
    ax.set(title=name, xticks=[], yticks=[])

fig.show()
```

```python
import numpy as np

x = np.arange(5)
y = np.arange(start=10, stop=15)

fig, axes = plt.subplots(nrows=1, ncols=2)
v_bar = axes[0].bar(x, y, color='lightblue', align='center')
h_bar = axes[1].barh(x, y, color='lightblue', align='center')
l1 = axes[0].axhline(10, color='gray', linewidth=2)
l2 = axes[1].axvline(10, color='gray', linewidth=2)

fig.show()
```

```python
fig, ax = plt.subplots()
v_bars = ax.bar(x, y, color='lightblue', align='center')
for bar in v_bars:
    bar.set(edgecolor='k', color='salmon', linewidth=3)
    ax.add_artist(bar)
fig.show()
```
